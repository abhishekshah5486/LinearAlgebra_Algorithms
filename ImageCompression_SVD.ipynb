```python
# Singular Value Decomposition SVD
```

```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
from google.colab import files
uploaded = files.upload()

image_path = next(iter(uploaded))
```

```python
color_image = imread(image_path)
plt.imshow(color_image)
```

```python
color_image.shape
```

```python
grayscale = np.mean(color_image, -1)
```

```python
grayscale.shape
```

```python
img = plt.imshow(grayscale)
img.set_cmap("gray")
plt.show()
```

```python
A = grayscale
U, S, Vt = np.linalg.svd(A, full_matrices = False)
S = np.diag(S)
print(U.shape, S.shape, Vt.shape)
print(S)
```

```python
def compress(r):
  x_approx = U[:, :r] @ S[:r, :r] @ Vt[:r, :]
  x_approx.shape
  approx_img = plt.imshow(x_approx)

  return approx_img

compressed_img = compress(20)
compressed_img.set_cmap('gray')
plt.show()

```
